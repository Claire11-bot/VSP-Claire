```python
from sklearn import linear_model
import pandas as pd
import numpy as np

#3 stream
monthly_counts_df = monthly_counts.reset_index()
monthly_counts_df.columns = ['month', 'event_count']
```

```python
# Convert month to integer (e.g. 202201, 202202, etc.)
monthly_counts_df['month'] = monthly_counts_df['month'].astype(str).apply(lambda x: int(x.replace('-', '')))

# Create a linear regression model
usa_reg = linear_model.LinearRegression()

X = monthly_counts_df[['month']]
y = monthly_counts_df['event_count']

usa_reg.fit(X, y)

beta = usa_reg.coef_[0]
alpha = usa_reg.intercept_

print(f"Beta: usa streams = {beta:.2f} month + {alpha:.2f}")
```

### USA vs ME countries

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assume GT_uae and GT_usa are data frames containing different data
df_uae = pd.DataFrame(GT_uae)
df_usa = pd.DataFrame(GT_us)
df_sa = pd.DataFrame(GT_sa)
df_qat = pd.DataFrame(GT_qat)

plt.figure(figsize=(10, 6))

# Draw a line chart of the first data set (UAE)
plt.plot(df_uae['palestina_United_Arab_Emirates'], color='blue', label='UAE')

# Draw a line chart of the second data set (USA)
plt.plot(df_usa['palestina_United_States'], color='red', label='USA')

# Draw a line chart of the first data set (SA)
plt.plot(df_sa['palestine_Saudi_Arabia'], color='orange', label='SA')

# Draw a line chart of the first data set (QAT)
plt.plot(df_qat['palestina_Qatar'], color='purple', label='QAT')


# Set chart title and axis labels
plt.title('Google Trends')
plt.xlabel('Week')
plt.ylabel('Trend Value')

# show grid
plt.grid(True)

# Set X-axis label rotation
plt.xticks(rotation=45)

# Show legend
plt.legend()

# display graphics
plt.tight_layout()
plt.show()
```
